# Exploratory Data Analysis

```python
sns.countplot(data=df, x="Vict_Sex", palette="pastel", order=df["Vict_Sex"].value_counts().index)
plt.title("Crime Rates by Victim Gender")
plt.xlabel("Victim Gender")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Location'].value_counts().head(10).plot(kind='bar', color='salmon')
plt.title("Top 10 Crime Locations")
plt.xlabel("Location")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Crm_Cd'].value_counts().head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 Crm_Cd')
plt.xlabel('Crm_Cd')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x=df['LON'],y=df['LAT'])
plt.title('Geographical Hotspots for Reported Crimes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Vict_Age'], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Victim Ages in Reported Crimes")
plt.xlabel("Victim Age")
plt.ylabel("Frequency")
plt.show()
```
